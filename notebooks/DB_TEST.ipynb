DM TEST

```python
from statsmodels.stats.diagnostic import acorr_ljungbox
from scipy import stats

def dm_test(e1, e2):
    d = e1**2 - e2**2  # loss differential
    n = len(d)
    mean_d = d.mean()
    acov = np.cov(d)  # variance
    dm_stat = mean_d / np.sqrt(acov / n)
    p_value = 2 * stats.norm.cdf(-abs(dm_stat))
    return dm_stat, p_value
```

```python
competitors = {
    "SARIMA": error_sarima,
    "XGBoost": error_xgb,
    "LSTM": error_lstm,
    "HW": error_hw,
    "Prophet": error_prophet
}

for name, e in competitors.items():
    stat, p = dm_test(error_rf, e)
    print(f"DM(RF vs {name}): stat={stat:.4f}, p={p:.4f}")
```

```python
def dm_test_mse(e1, e2):
    d = np.abs(e1) - np.abs(e2)  # loss differential
    n = len(d)
    mean_d = d.mean()
    acov = np.cov(d)  # variance
    dm_stat = mean_d / np.sqrt(acov / n)
    p_value = 2 * stats.norm.cdf(-abs(dm_stat))
    return dm_stat, p_value
```

```python
competitors = {
    "SARIMA": error_sarima,
    "XGBoost": error_xgb,
    "LSTM": error_lstm,
    "HW": error_hw,
    "Prophet": error_prophet
}

for name, e in competitors.items():
    stat, p = dm_test_mse(error_rf, e)
    print(f"DM(RF vs {name}): stat={stat:.4f}, p={p:.4f}")
```
