```python
import pandas as pd
import numpy as np
```

```python

# Load Excel file
xl0 = pd.ExcelFile(r"C:\Users\Top Prix\Downloads\Quarts Horaires 2023 (1).xlsx")
xl1=pd.ExcelFile(r"C:\Users\Top Prix\Downloads\QH Année 2024.xlsx")
xl2=pd.ExcelFile(r"C:\Users\Top Prix\Downloads\QH Année 2025.xlsx")
```

```python
all_months = []

for sheet in xl0.sheet_names:
    df = pd.read_excel(xl0, sheet_name=sheet, header=None)
    
    # skip header row, take only RIN rows (every 2 rows starting from row 1)
    rin_rows = df.iloc[1::2].reset_index(drop=True)
    
    dates = rin_rows.iloc[:, 0]          # column A = date
    rin_values = rin_rows.iloc[:, 2:]    # columns C onwards = 96 quarter-hours
    
    # melt from wide to long format
    rin_values.columns = range(96)
    rin_values['date'] = dates.values
    
    long = rin_values.melt(id_vars='date', var_name='quarter', value_name='reseau_rin_mw')
    long['datetime'] = pd.to_datetime(long['date']) + pd.to_timedelta(long['quarter'] * 15, unit='m')
    
    all_months.append(long[['datetime', 'reseau_rin_mw']])

# Combine all sheets
df_full0 = pd.concat(all_months, ignore_index=True)
df_full0 = df_full0.sort_values('datetime').set_index('datetime')


df_full0.head()
```

```python
all_months = []

for sheet in xl1.sheet_names:
    df = pd.read_excel(xl1, sheet_name=sheet, header=None)
    
    # skip header row, take only RIN rows (every 2 rows starting from row 1)
    rin_rows = df.iloc[1:-1,:].reset_index(drop=True)
    
    dates = rin_rows.iloc[:, 0]          # column A = date
    rin_values = rin_rows.iloc[:, 2:]    # columns C onwards = 96 quarter-hours
    
    # melt from wide to long format
    rin_values.columns = range(96)
    rin_values['date'] = dates.values
    
    long = rin_values.melt(id_vars='date', var_name='quarter', value_name='reseau_rin_mw')
    long['datetime'] = pd.to_datetime(long['date']) + pd.to_timedelta(long['quarter'] * 15, unit='m')
    
    all_months.append(long[['datetime', 'reseau_rin_mw']])

# Combine all sheets
df_full1 = pd.concat(all_months, ignore_index=True)
df_full1= df_full1.sort_values('datetime').set_index('datetime')


df_full1.head()
```

```python
#   Header row also has mixed/corrupt time values — skip it and read 96 cols from index 1
xl2 = pd.ExcelFile(r"C:\Users\Top Prix\Downloads\QH Année 2025.xlsx")
all_months = []
for sheet in xl2.sheet_names:
    df = pd.read_excel(xl2, sheet_name=sheet, header=None, nrows=32, usecols=range(97))
    data_rows  = df.iloc[1:].dropna(subset=[0]).reset_index(drop=True)  # skip header, drop empty rows
    dates      = data_rows.iloc[:, 0]                  # col A = date
    rin_values = data_rows.iloc[:, 1:97].copy()        # cols B–CQ = 96 QH values (NO label col)
    rin_values.columns = range(96)
    rin_values["date"] = dates.values
    long = rin_values.melt(id_vars="date", var_name="quarter", value_name="reseau_rin_mw")
    long["datetime"] = pd.to_datetime(long["date"]) + pd.to_timedelta(long["quarter"] * 15, unit="m")
    all_months.append(long[["datetime", "reseau_rin_mw"]])

df_full2 = pd.concat(all_months, ignore_index=True).sort_values("datetime").set_index("datetime")
print("2025:", df_full2.shape, df_full2.head(3))
```

```python
df_full=pd.concat([df_full0, df_full1, df_full2]).sort_index()
```

```python
df_full.to_csv("processed_rin_data.csv")
```
